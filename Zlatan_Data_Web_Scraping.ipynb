```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://fbref.com/en/players/4cde5509/Zlatan-Ibrahimovic#all_stats_standard'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
#table name
#<div id="all_stats_standard" class="table_wrapper tabbed">
```

```python
#assigned to a variable
table = soup.find("table")
```

```python
table.find_all('th')
```

```python
top_header = table.find_all('th')
print(top_header)
```

```python
titles = [title.text for title in top_header if title]
titles = [x for x in titles if x]
season = titles[titles.index('2001'):titles.index('22 Seasons')]
titles = titles[titles.index('Season'):titles.index('Matches')]
print(season)
print(titles)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = titles)
```

```python
df["Season"] = season
```

```python
table.find_all('tr')
```

```python
column_data = table.find_all('tr')
column_data
```

```python
column_values = []
for row in column_data:
    row_data = row.find_all('td')
    indi_row_data = [data.text for data in row_data]
    indi_row_data = indi_row_data[:len(indi_row_data)-1]
    column_values.append(indi_row_data)
column_values = column_values[2:27]
```

```python
for i in range(0, 25):
    for j in range (0, 34):
        df.iat[i, j+1] = column_values[i][j]
```

```python
df
```

```python
df.to_csv("Zlatan_Ibrahimovic.csv", index = False)
```
